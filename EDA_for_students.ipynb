```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"C:\Users\HP\Downloads\stud.csv"
df= pd.read_csv(file_path)
df.info()
df['parental_level_of_education']
```

```python
df.isnull().sum()
df.nunique()
```

```python
df['Total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['Average_score']=df['Total_score']/3
```

```python
df.describe()
df.head()
```

```python
reading_full=df[df['reading_score']==100]['Average_score'].count()

Writing_score=df[df['writing_score']==100]['Average_score'].count()
Maths_score=df[df['math_score']==100]['Average_score'].count()
print("total number of students score of 100 in maths ",Maths_score)
print("total number of students Score of 100 in writing ",Writing_score)
print("total number of students Score of 100 in reading ",reading_full)
```
